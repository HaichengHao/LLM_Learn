np.any()至少有一个满足条件,就返回true
np.all()所有的元素都满足条件,才返回true

```python
arr = np.array([[1,3,2,6,5],[8,6,4,9,0]])
```

```python
np.any(arr)
```

```python
np.any(arr>1)
```

```python
np.any(arr>100)
```

```python
np.all(arr>=0)
```

```python
np.all(arr>0)
```

```python
np.sort(arr) #tips:默认按行升序排序
```

```python
np.sort(arr,axis=0) #tips:按列升序排序
```

```python
np.sort(arr,axis=1) #tips：按行升序排序
```
