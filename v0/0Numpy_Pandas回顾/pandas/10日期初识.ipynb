```python
import pandas  as pd
```

```python
df = pd.DataFrame({'gmv':[100,200,300,400],'trade_date':["2025-01-06","2023-10-31","2023-12-31","2023-01-05"]})
df
```

```python
df['ymd']=pd.to_datetime(df['trade_date'])
```

```python
df
```

```python
df.dtypes
```

**dt是时间属性访问器对象**

```python
df["year"]=df["ymd"].dt.year
df
```

```python
df["ymd"].dt.month
```

```python
df["ymd"].dt.day
```

```python
df["ymd"].dt.dayofweek
```

```python
df['ymd'].dt.quarter #tips:查看季度
```

```python
df["ymd"].dt.is_month_end #tips:看看是不是月底
```

```python
df["ymd"].dt.is_year_end #tips:看看是不是年底
```

```python
df['ystat']=df['ymd'].dt.to_period("Y") #按年统计
df['mstat']=df['ymd'].dt.to_period('M') #按月统计
df['qstat']=df['ymd'].dt.to_period('Q') #按季度统计
df['wstat']=df['ymd'].dt.to_period('W') #按周统计
```
