```python
df = pd.DataFrame({"age":[20,22,18,10],"name":['niko','zzz','nox','ze'],"id":[101,102,103,104]})
df.set_index("id",inplace=True) #tips:让id作为列索引
df
```

```python
df.to_csv('./datas/demo.csv',encoding="gbk") #tips:如果想用excel这样的软件打开的话需要指定编码为gbk编码
df.to_json('./datas/demo.json',force_ascii=False) #tips:默认导出的是ascii,如果想要友好观察那就设置为force_ascii=False
```
