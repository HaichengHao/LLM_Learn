```python
import pandas as pd
```

## concat连接

```python
s1=pd.Series(index=["A","B"],data=[1,2])
s2=pd.Series(index=["D","E"],data=[1,2])
s3=pd.Series(index=["G","H"],data=[1,2])
pd.concat([s1,s2,s3],axis=0) #按照行连接，行的方向是垂直的,所以是一条向下扩展的
```

```python
pd.concat([s1,s2,s3],axis=1) #按列连接，不同列的方向是水平的,所以是一条向右的竖线
```

## DataFrame和Serise连接

```python
df1 = pd.DataFrame(data={"a": [1, 2], "b": [4, 5]}, index=[1, 2])
s1 = pd.Series(data=[7, 10], index=[1, 2], name="a")
```

```python
df1
```

```python
s1
```

```python
pd.concat([df1,s1]) #tips:默认按行连接，就是一行下面跟着一行,如果维度不匹配则填充Nan
```

```python
pd.concat([df1,s1],axis=1) #按照列连接
```

## dataframe与dataframe的连接

```python
df1 = pd.DataFrame(data={"a": [1, 2], "b": [4, 5]}, index=[1, 2])
df2 = pd.DataFrame(data={"a": [7, 8], "b": [10, 11]}, index=[1, 2])
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1)
```

## 重置索引

可通过ignore_index=True来重置索引。
重置索引之后索引会重新计算,而抛弃原来的索引

```python
df1 = pd.DataFrame(data={"a": [1, 2], "b": [4, 5]}, index=[1, 2])
df2 = pd.DataFrame(data={"a": [7, 8], "b": [10, 11]}, index=[1, 2])
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

## 类似join的连接

默认的合并方式是对其他轴进行并集合并（join=outer），可以用join=inner实现其他轴上的交集合并。

```python
df1 = pd.DataFrame(data={"a": [1, 2], "b": [4, 5]}, index=[1, 2])
df2 = pd.DataFrame(data={"b": [7, 8], "c": [10, 11]}, index=[2, 3])
```

```python
pd.concat([df1, df2],join="outer") #tips:join默认就是outer,也就是类似于sql中的全外连接
```

```python
pd.concat([df1,df2],join="inner") #tips:这样就实现了类似于sql的内连接操作
```
