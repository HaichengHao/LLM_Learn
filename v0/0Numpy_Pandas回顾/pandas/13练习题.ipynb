```python
df2 = pd.DataFrame({'Fruit':['Apple','Banana','Apple','Banana'],'Price':[2,1,3,1.5],'Quantity':[5,8,3,6]})
df2
```

```python
#使用groupby按照Fruit分组,计算每种水果的总销售额(销售额=价格*数量)

df2['Price']
```

```python
df2['Quantity']
```

```python
df2['total']=df2['Price']*df2['Quantity']
df2
```

```python
df2.groupby('Fruit')['total'].sum()
```
