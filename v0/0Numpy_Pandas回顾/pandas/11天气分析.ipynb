````python
df['month']=pd.to_datetime(df['date']).dt.to_period('M').astype(str) #tips:按月进行统计
````

## 分组聚合

```python
df.groupby('分组字段')["要聚合的的字段"].聚合函数()
df.groupby(['分组字段1','分组字段2',...])[["要聚合的字段1","要聚合的字段2"]].聚合函数()
```

````python
df.groupby('month')[["temp_max","temp_min"]].mean() #important:按月分组

#tips:按照sql理解 select AVG(temp_max),AVG(temp_min) from df groupby month;
````

````python
#important:统计每个月不同天气状况的数量

df.groupby('month')['weather'].nunique()

#tips:SELECT COUNT(DISTINCT weather) FROM df GROUP BY month;
````

查看每年最高温度的几种方式

````python
df['year_highest_temp']=pd.to_datetime(df['date']).dt.year
df.groupby('year_highest_temp')['temp_max'].max()
````
