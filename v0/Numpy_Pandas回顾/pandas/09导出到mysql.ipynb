```python
import pandas as pd
import os
from dotenv import load_dotenv
db_pwd = os.getenv('mysqlpwd')
```

```python
df = pd.DataFrame({"age":[20,22,18,10],"name":['niko','zzz','nox','ze'],"id":[101,102,103,104]})
df.set_index("id",inplace=True) #tips:让id作为列索引
df
```

```python
from sqlalchemy import create_engine
engine=create_engine(
    f"mysql+pymysql://root:{db_pwd}@localhost:3306/df2sql"
)
```

```python
df.to_sql(
    name="demo2sql",
    con=engine,
    if_exists="replace",
    index=False, #tips:不将DataFrame的索引写入数据库
    chunksize=1000,#tips:批量插入的大小,对于大数据集很有用
)
print('数据成功导出到数据库')
```

## 拿到数据库中的数据

```python
df2 = pd.read_sql(
    "select * from demo2sql",
    con=engine
)
df2
```
