```python
import numpy as np
from numpy import dtype
from tornado.autoreload import start
```

```python
data = [1,2,3]
print(f'data的内存地址{id(data)}')

arr1=np.array(data) #从数组列表创建arry tips: 包解释 create an arry
print(arr1)
print(type(arr1))
print(f"arr1的内存地址{id(arr1)}")
```

```python
arr2 = np.asarray(data)  #tips:包解释是Convert the input to an array.
print(arr2)
print(type(arr2))
print(f"arr2的内存地址{id(arr2)}")
```

```python
arr3 = np.asarray(arr2)
print(arr3)
print(type(arr3))
print(f"arr3的内存地址{id(arr3)}")
```

这也就验证了猜想,也就是说asarry不会改变原来的内存地址,而只是对原来的进行了转化,但是arry则会新建一个新的ndarry并开辟新的内存地址
