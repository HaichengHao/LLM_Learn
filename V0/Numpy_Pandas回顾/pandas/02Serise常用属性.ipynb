loc[] 显式索引,按标签索引或切片
iloc[]隐式索引,按位置索引或切片
at[]使用标签访问单个元素
iat[]使用位置访问单个元素

```python
s1.loc['a':'c']
#important:从a-c索引,闭区间
```

```python
s1.iloc[0:2]
#important:从索引标0-2,开区间
```

```python
s1.iloc[0:3]
```

```python
s1.at['c']
```
