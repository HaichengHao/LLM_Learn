```python
from langchain_community.document_loaders import UnstructuredWordDocumentLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
docs = UnstructuredWordDocumentLoader(
    file_path='../assets/sample.docx',
    mode='single'
).load()
```

只有一个元素可不是我们想要的,所以接下来是本节的开头,也就是进行切分,将这个列表中的0号元素(本来也就只有它一个哈)切成多个,并返回一个新的列表

```python
#important:切分为文本块儿

splitter=RecursiveCharacterTextSplitter(
    separators=['\n\n','\n','。','！','？','……','，',''], #tips:分割符列表,按顺序切分，现用句号切分，然后是感叹号，然后问号,然后省略号，然后逗号，然后空
    chunk_size=400, #每个块儿的最大长度
    chunk_overlap=50,#tips:每个块重叠的长度
    length_function=len,#tips:可选,计算文本长度的函数，默认为字符串长度,可以自定义函数来实现token数切分
    add_start_index=True #tips:可选,块的元数据中添加如此块起始索引

)
splitted_res = splitter.split_documents(docs)
```

api版本,用openai的模型

```python
from langchain_openai import OpenAI
from dotenv import load_dotenv
import os
load_dotenv()

api_key = os.getenv('api_key')
base_url = os.getenv('base_url')
```

```python
from langchain_openai import OpenAIEmbeddings

embedd_model = OpenAIEmbeddings(
    api_key=api_key,
    base_url=base_url,
    model="text-embedding-3-small",
    dimensions=128 #tips:指定维度,不然可能默认的太大就会有点慢
)
demoquery='你好，世界'
demodocs=['你好','世界','人工智能','未来']
print(embedd_model.embed_query(demoquery))
print(embedd_model.embed_documents(demodocs))
```
