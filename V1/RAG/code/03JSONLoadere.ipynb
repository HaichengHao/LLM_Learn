````python
from langchain_community.document_loaders import JSONLoader
````

````python
loader = JSONLoader(
    file_path='../assets/sample.json',
    jq_schema=".", #important:jq_schema是提取json当中对象的一种特定方式,特定的语法,例如"."就是提取整个的json对象,其实可以理解为字典那样的点取值法
    text_content=False
)
````

````python
load_document = loader.load()
load_document
````

下面再构建一个loader2,并指定一下jq_schema,譬如我们只要status

````python
loader2 = JSONLoader('../assets/sample.json',jq_schema='.status', text_content=False)
load_document2 = loader2.load()
load_document2
````

那么如果想要data中的total_items呢?

````python
loader3 = JSONLoader('../assets/sample.json',jq_schema='.data.total_items', text_content=False)
load_document3 = loader3.load()
load_document3
````

现在彻底明白了jq_schema的作用了,来个更复杂的,要data下的items下的第二个元素的author的name

````python
loader4 = JSONLoader('../assets/sample.json',jq_schema='.data.items[1].author.name',text_content=False)
load_document4 = loader4.load()
load_document4
````

那如果是拿到要data下的items下的第所有元素的author的name呢

````python
loader5 = JSONLoader('../assets/sample.json',jq_schema='.data.items[].author.name',text_content=False) #tips:其实很简单,不作具体指定是哪个元素就行了
load_document5 = loader5.load()
load_document5
````

那么后面这个text_content什么时候可以是True呢?就像上面我们只要status的单个“success"字符串输出的时候而不是结构性输出(譬如字典列表...)的时候就可以让text_content为True了

````python
loader6 = JSONLoader('../assets/sample.json',jq_schema='.status', text_content=True)
load_document6 = loader6.load()
load_document6
````
