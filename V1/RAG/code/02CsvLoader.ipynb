还有一种玩法,我们可以指定content_columns,这样Loader就会只将我们指定的列作为内容了

```python
loader2 = CSVLoader(file_path='../assets/sample.csv',encoding='utf-8',content_columns=['content'])
```

```python
load_document2=loader2.load()
load_document2
```

```python
load_document2[0].metadata
```

```python
load_document2[0].page_content
```

```python
loader3 = CSVLoader(
    file_path='../assets/sample.csv',
    encoding='utf-8',
    metadata_columns=['title','author'], #tips:将指定列作为元数据
    content_columns=['content'] #tips:将指定列作为内容数据
)
load_document3 =loader3.load()
```

```python
load_document3
```

```python
load_document3[0].metadata
```
