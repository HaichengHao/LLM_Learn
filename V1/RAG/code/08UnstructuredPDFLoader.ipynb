````python
from langchain_community.document_loaders import UnstructuredPDFLoader
````

````python
loader=UnstructuredPDFLoader(
    file_path='../assets/sample.pdf',
    mode='elements',
    #single:返回单个document对象
    #elements:按标题等元素切分文档
    strategy='hi_res',
    #加载策略
    #fast:pdfminer 提取并处理文本
    #ocr_only:转化为图片并进行OCR
    #hi_res:识别文档布局，将OCR输出与pdfminer输出融合

    infer_table_structure=True,
    #OCR推断表格结构,仅hi_res下起效，如果为True则会在表格元素的元数据中添加text_as_html

    language=['eng','chi_smi']

    #OCR使用的语言，eng英文，chi_smi简中 https://github.com/tesseract-ocr/langdata


)

loaddoc=loader.load()
loaddoc
````
